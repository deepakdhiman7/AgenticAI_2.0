```python
import os
from dotenv import load_dotenv
from pinecone import Pinecone, ServerlessSpec
# from langchain_huggingface.embeddings import HuggingFaceEmbeddings
from langchain_google_genai.embeddings import GoogleGenerativeAIEmbeddings
```

```python
load_dotenv()

os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["HF_TOKEN"]=os.getenv("HF_TOKEN")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
# os.environ["PINECODE_API_KEY"]=os.getenv("PINECODE_API_KEY")
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
```

```python
# hf_embedding = HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
google_embedding = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")
```

```python
embedded_text = google_embedding.embed_query("my name is deepak")
len(embedded_text)
```

```python
# creating Pinecone client
pc = Pinecone(api_key=os.getenv("PINECODE_API_KEY"))
```

```python
# creating the index
index_name = "agentic-v2"

if pc.has_index(index_name):
    print("index already existing...")
else:
    pc.create_index(
        name=index_name,
        spec=ServerlessSpec(cloud="aws", region="us-east-1"),
        dimension=768,
        metric="cosine",
    )
```

```python
# loading the index
index = pc.Index(index_name)
```
