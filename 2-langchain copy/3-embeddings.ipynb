# Embeddings

```python
import os
from dotenv import load_dotenv
from langchain_openai import OpenAIEmbeddings
from sentence_transformers import SentenceTransformer
```

```python
load_dotenv()

# print(os.getenv("OPENAI_API_KEY"))
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
```

```python
embedding = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
query = "this is agentic ai course. my name is deepak"
```

```python
result = embedding.embed_query(query)
```

```python
print(type(result))
print(len(result))
# result
```

```python
result[0:3]
```

## using sentence transformers embedding

```python
sentence_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sentence_embedding = sentence_model.encode(query)
```

```python
print(type(sentence_embedding))
print(len(sentence_embedding))
# print(sentence_embedding)
```
